```
#get seperation for comparison
mobileData=train[train.advertiser==2997]
milkData=train[train.advertiser==2259]
mobileData=mobileData.sort_values(by='region')
```

```
#ECPC Weekday:
mobile = mobileData[['payprice','weekday']]
milk = milkData[['payprice','weekday']]
#COUNT
i=0
while i < 7:
  Click1 = milk[milk.weekday==i]
  sum = Click1['payprice'].sum()
  print(str(i)+ ":"+str(sum))
  i=i+1
```

```
#ECPC hour:
mobile = mobileData[['payprice','hour']]
milk = milkData[['payprice','hour']]
#COUNT
i=0
while i < 24:
  Click1 = milk[milk.hour==i]
  sum = Click1['payprice'].sum()
  print(str(i)+ ":     "+str(sum))
  i=i+1
```

```
#ECPC region:
mobile = mobileData[['payprice','region']]
milk = milkData[['payprice','region']]
#COUNT
i=0
while i < 395:
  Click1 = mobile[mobile.region==i]
  sum = Click1['payprice'].sum()
  if sum != 0:
    print(str(i)+ ":     "+str(sum))
  i=i+1
```
